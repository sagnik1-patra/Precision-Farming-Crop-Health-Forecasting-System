```python
# ============================================================
# 🌾 Hybrid Precision Farming & Crop Health Forecasting System
# Author : Annan Sadr
# Hybrid Model = AIS (Artificial Immune System) + GA (Genetic Algorithm)
# ============================================================

import os
import json
import yaml
import random
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import joblib

# ------------------------------------------------------------
# 📂 Paths
# ------------------------------------------------------------
BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Precision Farming & Crop Health Forecasting System"
PATH_PEST = os.path.join(BASE_DIR, "archive", "pesticides.csv")
PATH_RAIN = os.path.join(BASE_DIR, "archive", "rainfall.csv")
PATH_TEMP = os.path.join(BASE_DIR, "archive", "temp.csv")
PATH_YIELD = os.path.join(BASE_DIR, "archive", "yield.csv")

# ------------------------------------------------------------
# 🧩 Helper Functions
# ------------------------------------------------------------
def load_csv_optimized(path, name):
    df = pd.read_csv(path, low_memory=False)
    print(f"  → Loaded {name}: {df.shape}")
    for col in df.select_dtypes(include=["float64", "int64"]).columns:
        df[col] = pd.to_numeric(df[col], downcast="float")
    return df

def find_year_column(df):
    for c in df.columns:
        if str(c).strip().lower() == "year":
            return c
    raise ValueError(f"No 'Year' column found in: {df.columns.tolist()}")

def aggregate_yearly(df):
    year_col = find_year_column(df)
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    grouped = df.groupby(year_col, as_index=False)[numeric_cols].mean()
    grouped.rename(columns={year_col: "Year"}, inplace=True)
    return grouped

# ------------------------------------------------------------
# 🧩 Load & Merge Datasets
# ------------------------------------------------------------
print("[INFO] Loading datasets...")
df_pest = load_csv_optimized(PATH_PEST, "Pesticides")
df_rain = load_csv_optimized(PATH_RAIN, "Rainfall")
df_temp = load_csv_optimized(PATH_TEMP, "Temperature")
df_yield = load_csv_optimized(PATH_YIELD, "Yield")

print("[INFO] Aggregating yearly data...")
pest_yearly = aggregate_yearly(df_pest)
rain_yearly = aggregate_yearly(df_rain)
temp_yearly = aggregate_yearly(df_temp)
yield_yearly = aggregate_yearly(df_yield)

df_merged = (
    yield_yearly.merge(pest_yearly, on="Year", how="left")
                .merge(rain_yearly, on="Year", how="left")
                .merge(temp_yearly, on="Year", how="left")
)
df_merged = df_merged.dropna().reset_index(drop=True)
print(f"[INFO] Final merged shape: {df_merged.shape}")

# ------------------------------------------------------------
# 🎯 Features & Target
# ------------------------------------------------------------
target_col = "Yield" if "Yield" in df_merged.columns else df_merged.columns[-1]
feature_cols = [c for c in df_merged.columns if c not in ["Year", target_col]]
X = df_merged[feature_cols].values
y = df_merged[target_col].values.reshape(-1, 1)

scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()
X_scaled = scaler_x.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

# ------------------------------------------------------------
# 🧬 Hybrid AIS + GA Optimizer
# ------------------------------------------------------------
def create_model(n1, n2, lr):
    model = Sequential([
        Dense(n1, activation="relu", input_dim=X_train.shape[1]),
        Dropout(0.2),
        Dense(n2, activation="relu"),
        Dense(1, activation="linear")
    ])
    model.compile(optimizer=Adam(learning_rate=lr), loss="mse")
    return model

def fitness(params):
    n1, n2, lr = params
    model = create_model(int(n1), int(n2), lr)
    model.fit(X_train, y_train, epochs=50, batch_size=4, verbose=0)
    loss = model.evaluate(X_test, y_test, verbose=0)
    return -loss  # maximize fitness (lower loss = higher fitness)

# AIS + GA hybrid search
def hybrid_optimize(pop_size=8, generations=5):
    population = [[random.randint(16,128), random.randint(8,64), random.uniform(0.0005,0.01)] for _ in range(pop_size)]
    for gen in range(generations):
        scores = [fitness(p) for p in population]
        best_idx = np.argmax(scores)
        best = population[best_idx]
        print(f"[AIS+GA] Gen {gen+1}: Best -> n1={best[0]}, n2={best[1]}, lr={best[2]:.5f}, Fitness={scores[best_idx]:.4f}")
        # Selection + mutation
        new_pop = [best.copy()]
        while len(new_pop) < pop_size:
            p = best.copy()
            if random.random() < 0.5: p[0] += random.randint(-8, 8)
            if random.random() < 0.5: p[1] += random.randint(-4, 4)
            if random.random() < 0.5: p[2] *= random.uniform(0.8, 1.2)
            p[0] = max(8, min(128, p[0]))
            p[1] = max(4, min(64, p[1]))
            p[2] = max(0.0001, min(0.02, p[2]))
            new_pop.append(p)
        population = new_pop
    return best

print("[INFO] Running Hybrid AIS + GA optimization...")
best_params = hybrid_optimize()
n1_opt, n2_opt, lr_opt = best_params
print(f"[INFO] Optimal Params -> n1={n1_opt}, n2={n2_opt}, lr={lr_opt:.5f}")

# ------------------------------------------------------------
# 🧠 Train Final Model
# ------------------------------------------------------------
model = create_model(int(n1_opt), int(n2_opt), lr_opt)
history = model.fit(X_train, y_train, epochs=200, batch_size=4, validation_split=0.2, verbose=0)

y_pred_scaled = model.predict(X_test)
y_pred = scaler_y.inverse_transform(y_pred_scaled)
y_true = scaler_y.inverse_transform(y_test)
rmse = np.sqrt(np.mean((y_true - y_pred) ** 2))
mae = np.mean(np.abs(y_true - y_pred))
r2 = 1 - (np.sum((y_true - y_pred) ** 2) / np.sum((y_true - np.mean(y_true)) ** 2))

# ------------------------------------------------------------
# 💾 Save Results & Graphs
# ------------------------------------------------------------
joblib.dump(scaler_x, os.path.join(BASE_DIR, "hybrid_scaler_x.pkl"))
joblib.dump(scaler_y, os.path.join(BASE_DIR, "hybrid_scaler_y.pkl"))
model.save(os.path.join(BASE_DIR, "hybrid_precision_model.h5"))

results = {"RMSE": float(rmse), "MAE": float(mae), "R2_Score": float(r2),
           "n1": n1_opt, "n2": n2_opt, "lr": lr_opt}
with open(os.path.join(BASE_DIR, "hybrid_precision_results.json"), "w") as f:
    json.dump(results, f, indent=4)

plt.style.use("seaborn-v0_8-darkgrid")

# 1️⃣ Accuracy Graph
plt.figure(figsize=(8,5))
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Val Loss")
plt.title("Hybrid AIS+GA Model Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(BASE_DIR, "hybrid_precision_accuracy_graph.png"))
plt.show()

# 2️⃣ Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df_merged[feature_cols + [target_col]].corr(), annot=True, cmap="coolwarm")
plt.title("Hybrid Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig(os.path.join(BASE_DIR, "hybrid_precision_heatmap.png"))
plt.show()

# 3️⃣ Comparison Graph
plt.figure(figsize=(8,5))
plt.scatter(y_true, y_pred, color="blue", alpha=0.7)
plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], "r--")
plt.title("Hybrid Actual vs Predicted Yield")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.tight_layout()
plt.savefig(os.path.join(BASE_DIR, "hybrid_precision_comparison_graph.png"))
plt.show()

# 4️⃣ Prediction Graph
plt.figure(figsize=(8,5))
years = df_merged["Year"][-len(y_pred):]
plt.plot(years, y_pred.flatten(), "b-o", label="Predicted Yield")
plt.title("Hybrid Predicted Yield Trend by Year")
plt.xlabel("Year")
plt.ylabel("Predicted Yield")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(BASE_DIR, "hybrid_precision_prediction_graph.png"))
plt.show()

# 5️⃣ Result Graph
plt.figure(figsize=(6,4))
metrics = ["RMSE", "MAE", "R²"]
values = [rmse, mae, r2]
bars = plt.bar(metrics, values, color=["#4CAF50","#2196F3","#FFC107"])
plt.title("Hybrid Model Performance")
for bar in bars:
    plt.text(bar.get_x()+bar.get_width()/2., bar.get_height(), f"{bar.get_height():.3f}", ha="center", va="bottom")
plt.tight_layout()
plt.savefig(os.path.join(BASE_DIR, "hybrid_precision_result_graph.png"))
plt.show()

# ------------------------------------------------------------
# 🧾 Summary
# ------------------------------------------------------------
print("\n✅ HYBRID AIS+GA MODEL TRAINED SUCCESSFULLY!")
print(f"RMSE: {rmse:.3f} | MAE: {mae:.3f} | R²: {r2:.3f}")
print("[INFO] All hybrid results saved in:", BASE_DIR)
```
